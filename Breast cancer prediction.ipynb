```python
# import libraries
import pandas as pd # for data manupulation or analysis
import numpy as np # for numeric calculation
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns # for data visualization
```

```python
#Load breast cancer dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
type(cancer_dataset)
```

```python
# keys in dataset
cancer_dataset.keys()
```

```python
# features of each cells in numeric format
cancer_dataset['data']
```

```python
# malignant or benign value
cancer_dataset['target']
```

```python
# name of features
print(cancer_dataset['feature_names'])
```

```python
# create datafrmae
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
             columns = np.append(cancer_dataset['feature_names'], ['target']))
cancer_df.head()
```

```python
# Information of cancer Dataframe
cancer_df.info()
```

```python
# Numerical distribution of data (stats)
cancer_df.describe()
```

```python
# Count the target class
sns.countplot(cancer_df['target'])
```

```python
#Split dataframe in train and set
# input variable
X = cancer_df.drop(['target'], axis = 1)
X.head()
```

```python
# output variable
y = cancer_df['target']
y.head()
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
# supervised classification machine learning algorithms.
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Support vector classifier
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
y_pred_scv = svc_classifier.predict(X_test)
accuracy_score(y_test, y_pred_scv)
```

```python
# Train with Standard scaled Data
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_svc_sc)
```

```python
print("Accuracy score %f" % accuracy_score(y_test, y_pred_svc_sc))
print(classification_report(y_test, y_pred_svc_sc))
```
